```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#usando o arquivo missing_data_per_snp.lmiss
 #SNP                SNP identifier
 #CHR                Chromosome number
 #N_MISS             Number of individuals missing this SNP
 #N_GENO             Number of non-obligatory missing genotypes
 #F_MISS             Proportion of sample missing for this SNP
```

```python
miss_data = pd.read_csv('missing_data_per_snp.csv')
number_of_snps = len(miss_data)
number_miss_data = miss_data['F_MISS'].values
```

```python
fig, [ax, ax1] = plt.subplots(1,2, figsize=(10,4))
ax.hist(number_miss_data, histtype='barstacked', rwidth=0.8, facecolor='r', edgecolor="black", log=True)
ax.set_ylabel("Number os SNPs (log scale)")
ax.set_xlabel("Number of Missing Genotyping Data")
ax.set_yticks((10, 100, 1000, 7000),(10**1, 10**2, 10**3, 7000))

ax1.hist(number_miss_data, histtype='barstacked', rwidth=0.8, facecolor='r', edgecolor="black")
ax1.set_ylabel("Number os SNPs")
ax1.set_xlabel("Number of Missing Genotyping Data")
```
