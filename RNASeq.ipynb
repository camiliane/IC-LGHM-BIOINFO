1. carregar cada pasta em um for loop
2. importar somente os arquivos de mrna
3. carregar todos como pandas dataframe (read_csv)

```python
rsem = ["luad_tcga", "luad_tcga_pan_can_atlas_2018", "luad_tcga_pub", "lusc_tcga","lusc_tcga_pan_can_atlas_2018", 
        "paad_cptac_2021", "paad_qcmg_uq_2016", "paad_tcga", "paad_tcga_pan_can_atlas_2018", "stad_tcga",
        "stad_tcga_pan_can_atlas_2018", "stad_tcga_pub"]

dfs_rsem = []
dfs_rpkm = []
dfs_ = []
for filename in files:
    path = os.path.join(directory, filename)

    rnaseq_rsem = glob.glob(path + "\\data_mrna_seq_v2_rsem.txt")
    rnaseq_rpkm = glob.glob(path + "\\data_mrna_seq_rpkm.txt") 
    
    for file in rnaseq_rsem:
        df_rsem = pd.read_csv(file, sep="\t").T
        new_header = df_rsem.iloc[0] 
        df_rsem = df_rsem[1:] 
        df_rsem.columns = new_header
        #df_rsem['study']='{}'.format(rsem[for i in len(12)])
        df_rsem['study'] = ['{}'.format(rsem[i]) for i in range(len(rsem))]
        dfs_rsem.append(df_rsem)
```
